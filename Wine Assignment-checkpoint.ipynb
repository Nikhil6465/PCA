```python
wine_ary=wine_data.values
wine_ary
```

```python
scaled = StandardScaler()
wine_scaled = scaled.fit_transform(wine_ary)
wine_scaled
```

```python
pca=PCA(n_components=13)

wine_pca=pca.fit_transform(wine_scaled)
wine_pca
```

```python
pca.components_
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
plt.plot(var1,color='blue')
plt.show()
```

```python
final_df=pd.concat([wine['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=final_df)
plt.show()
```

```python
fig=plt.figure(figsize=(16,12))
sns.kdeplot(data=final_df)
plt.show()
```

### Using Clustering Algorithms

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(final_df,method='complete'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(final_df),columns=['clustersing'])
y['clustersing'].value_counts()
```

```python
final_df1=final_df.copy()
final_df1['clustersing']=hclusters.labels_
final_df1
```

## KMean clustring

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df_norm=StandardScaler().fit_transform(final_df)
df_norm
```

```python
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clusters4=KMeans(4,random_state=30).fit(df_norm)
clusters4
```

```python
clusters4.labels_
```

```python
df_norm1=final_df.copy()
df_norm1['clustersing']=clusters4.labels_
df_norm1
```

```python
clusters4.cluster_centers_
```

```python
df_norm1.groupby('clustersing').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10, 7))  
sns.scatterplot(df_norm1['clustersing'],df_norm1['Type'], c=clusters4.labels_) 
plt.show()
```
